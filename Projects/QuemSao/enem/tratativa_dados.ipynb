```python
#importaÃ§Ãµes e caminho
import pandas as pd

caminho_arquivo = "/content/PARTICIPANTES_2024.csv"
#leitura do csv
df = pd.read_csv(caminho_arquivo, sep=';', encoding='latin-1', low_memory=False)
print(df.columns.tolist())
```

```python
#informaÃ§Ãµes gerais
print("DimensÃµes:", df.shape)
print("\nColunas:", df.columns.tolist())
print("\nTipos:", df.dtypes)
print("\n% de valores ausentes:")
print(df.isnull().mean().round(3) * 100)
print("\nEstatÃ­sticas descritivas (numÃ©ricas):")
print(df.describe())
print("\nPrimeiras 5 linhas:", df.head())
```

```python
#dicionario colunas
novo_nome_colunas = {
    'NU_INSCRICAO': 'inscricao',
    'NU_ANO': 'ano',
    'TP_FAIXA_ETARIA': 'faixa_etaria',
    'TP_SEXO': 'sexo',
    'TP_ESTADO_CIVIL': 'estado_civil',
    'TP_COR_RACA': 'cor_raca',
    'TP_NACIONALIDADE': 'nacionalidade',
    'TP_ST_CONCLUSAO': 'status_conclusao',
    'TP_ANO_CONCLUIU': 'ano_conclusao',
    'TP_ENSINO': 'tipo_ensino',
    'IN_TREINEIRO': 'treineiro',
    'CO_MUNICIPIO_PROVA': 'cod_municipio_prova',
    'NO_MUNICIPIO_PROVA': 'municipio_prova',
    'CO_UF_PROVA': 'cod_uf_prova',
    'SG_UF_PROVA': 'uf_prova',
    'Q001': 'esc_mae',
    'Q002': 'esc_pai',
    'Q003': 'trab_mae',
    'Q004': 'trab_pai',
    'Q005': 'pessoas_residencia',
    'Q006': 'possui_renda',
    'Q007': 'renda_mensal_familiar',
    'Q008': 'contrata_servico_domestico',
    'Q009': 'tem_banheiro',
    'Q010': 'tem_quarto',
    'Q011': 'tem_carro',
    'Q012': 'tem_motocicleta',
    'Q013': 'tem_geladeira',
    'Q014': 'tem_freezer',
    'Q015': 'tem_maquina_de_lavar',
    'Q016': 'tem_micro_ondas',
    'Q017': 'tem_aspirador_de_po',
    'Q018': 'tem_tv',
    'Q019': 'tem_tv_por_assinatura',
    'Q020': 'tem_internet_wifi',
    'Q021': 'tem_computador_notebook',
    'Q022': 'tem_celular',
    'Q023': 'tipo_escola'
}

colunas_renomeadas = [
    'inscricao', 'ano', 'faixa_etaria', 'sexo', 'estado_civil', 'cor_raca', 'nacionalidade',
    'status_conclusao', 'ano_conclusao', 'tipo_ensino', 'treineiro', 'cod_municipio_prova',
    'municipio_prova', 'cod_uf_prova', 'uf_prova', 'esc_mae', 'esc_pai', 'trab_mae', 'trab_pai',
    'pessoas_residencia', 'possui_renda', 'renda_mensal_familiar', 'contrata_servico_domestico',
    'tem_banheiro', 'tem_quarto', 'tem_carro', 'tem_motocicleta', 'tem_geladeira', 'tem_freezer',
    'tem_maquina_de_lavar', 'tem_micro_ondas', 'tem_aspirador_de_po', 'tem_tv',
    'tem_tv_por_assinatura', 'tem_internet_wifi', 'tem_computador_notebook', 'tem_celular',
    'tipo_escola'
]

#renomear colunas
df = df.rename(columns=novo_nome_colunas)
df.head()
```

```python
#filtrar participantes da paraÃ­ba
df_pb_enem_2024 = df[df["uf_prova"] == "PB"]

#selecionar colunas de interesse
df_pb_enem_2024 = df_pb_enem_2024[colunas_renomeadas]
df_pb_enem_2024.head()
```

```python
#tratamento de valores ausentes
# Preenche objetos com string padrÃ£o
for col in df_pb_enem_2024.select_dtypes(include='object').columns:
    df_pb_enem_2024[col] = df_pb_enem_2024[col].fillna('NÃ£o informado')

# Preenche numÃ©ricos com mediana
for col in df_pb_enem_2024.select_dtypes(exclude='object').columns:
    df_pb_enem_2024[col] = df_pb_enem_2024[col].fillna(df_pb_enem_2024[col].median())

df_pb_enem_2024.head()
```

```python
#salvar o csv tratado
df_pb_enem_2024.to_csv("enem_pb_2024_tratado.csv", index=False)
print("Arquivo salvo com sucesso!")
```
