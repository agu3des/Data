## 8. Tratamento de Valores Ausentes (Missing Values)

Valores ausentes podem comprometer a análise.

### Estratégias de Tratamento:

1. **Variáveis Categóricas (texto)**: Preenchemos com `"Não informado"`
   - Exemplos: `sexo`, `tipo_escola`, `renda_mensal_familiar`
   - Justificativa: Preserva a informação de que o dado não foi fornecido

2. **Variáveis Numéricas**: Preenchemos com a **mediana** da coluna
   - Exemplos: `faixa_etaria`, `pessoas_residencia`
   - Justificativa: A mediana é robusta a outliers e representa o valor central

### Por que a mediana e não a média?

- A mediana é menos afetada por valores extremos
- Em distribuições assimétricas (comum em dados socioeconômicos), a mediana representa melhor o valor "típico"

```python
# Verificando valores ausentes antes do tratamento
print("="*60)
print("VALORES AUSENTES ANTES DO TRATAMENTO")
print("="*60)
missing_before = df_pb_enem_2024.isnull().sum()
missing_before_percent = (df_pb_enem_2024.isnull().mean() * 100).round(2)

missing_info = pd.DataFrame({
    'Coluna': missing_before.index,
    'Valores Ausentes': missing_before.values,
    'Percentual (%)': missing_before_percent.values
})

# Mostra apenas colunas com valores ausentes
missing_info_filtered = missing_info[missing_info['Valores Ausentes'] > 0]
if len(missing_info_filtered) > 0:
    print(missing_info_filtered.to_string(index=False))
else:
    print(" Nenhum valor ausente encontrado!")
```

```python
# Tratamento de valores ausentes

# 1. Variáveis categóricas (objetos): preencher com "Não informado"
print("\nPreenchendo valores ausentes em variáveis categóricas...")
colunas_categoricas = df_pb_enem_2024.select_dtypes(include='object').columns
for col in colunas_categoricas:
    qtd_ausentes = df_pb_enem_2024[col].isnull().sum()
    if qtd_ausentes > 0:
        df_pb_enem_2024[col] = df_pb_enem_2024[col].fillna('Não informado')
        print(f"  - {col}: {qtd_ausentes} valores preenchidos com 'Não informado'")

# 2. Variáveis numéricas: preencher com a mediana
print("\nPreenchendo valores ausentes em variáveis numéricas...")
colunas_numericas = df_pb_enem_2024.select_dtypes(exclude='object').columns
for col in colunas_numericas:
    qtd_ausentes = df_pb_enem_2024[col].isnull().sum()
    if qtd_ausentes > 0:
        mediana = df_pb_enem_2024[col].median()
        df_pb_enem_2024[col] = df_pb_enem_2024[col].fillna(mediana)
        print(f"  - {col}: {qtd_ausentes} valores preenchidos com mediana ({mediana})")
```

```python
# Verificando se ainda existem valores ausentes
print("\n" + "="*60)
print("VERIFICAÇÃO FINAL DE VALORES AUSENTES")
print("="*60)
valores_ausentes_final = df_pb_enem_2024.isnull().sum().sum()

if valores_ausentes_final == 0:
    print(" Nenhum valor ausente no dataset!")
    print(" Dados prontos para análise!")
else:
    print(f"Ainda existem {valores_ausentes_final} valores ausentes.")
    print("\nDetalhamento:")
    print(df_pb_enem_2024.isnull().sum()[df_pb_enem_2024.isnull().sum() > 0])
```
